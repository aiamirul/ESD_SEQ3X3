```python
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
# Load the YOLOv8 model (you can specify 'yolov8n', 'yolov8s', etc.)
# model = YOLO('/run/user/1000/gvfs/sftp:host=192.168.1.3/media/infer/SSDC/computer_vision/production_models/FRANCE/FRV/FRV_AUX_ACTIONSEQ_001/weights/best.pt')  # Replace with the model you want to use
model = YOLO('/run/user/1000/gvfs/sftp:host=192.168.1.3/media/infer/SSDC/computer_vision/production_models/FRANCE/FRV/FRV_AUX_ESD159OBJ_000/weights/best.pt')
# Path to the input video
video_path = "/media/amir/SSD_B/FR/FRV/001/2024082600107.mp4"
output_path = "output_video.mp4"  # Path to save the processed video
# Open the video file
cap = cv2.VideoCapture(video_path)

# Get video properties
fps = int(cap.get(cv2.CAP_PROP_FPS))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for output video

# Initialize video writer for saving the output
out = cv2.VideoWriter(output_path, fourcc, fps, (width, height))

# Loop through video frames
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Break when video ends

    # Perform inference on the frame
    results = model(frame)

    # Get annotated frame
    annotated_frame = results[0].plot()

    # Write the frame to the output video
    out.write(annotated_frame)

    # Display the frame (optional, for real-time viewing)
    cv2.imshow('YOLOv8 Inference', annotated_frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):  # Press 'q' to exit
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Processed video saved at {output_path}")
```
